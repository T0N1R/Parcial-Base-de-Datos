# Ejercicio 1

### ¿Qué es una base de datos?

Una base de datos es un conjunto de información (ya sea digital o no) almacenada y ordenada para su revisión o consulta. Su objetivo es poder realizar una consulta de información específica en el menor tiempo posible y realizar análisis para mejorar el rendimiento

### ¿Qué  es el proceso de Normalización?

Es el proceso de organización de las bases de datos para eliminar anomalías como información redundante, anomalías de actualización, problemas de insersión y proteger la integridad de la información de la base de datos.

### ¿Cuál es el objetivo principal de la Inteligencia de Negocios? ¿Cómo se relaciona este concepto con el concepto de DataWareHouse? 

El objetivo de Inteligencia de Negocios es la toma de decisiones en base a la retroalimentación obtenida de análisis de información recolectada de la empresa con el fin de reducir riesgos. La inteligencia de Negocios se relaciona con el concepto de DataWareHouse ya que una de las características principales de una buena inteligencia de negocios es tener fuentes de información confiable con el menor cesgo posible. Los Data Warehouse, además de la obtención de variasdatos de información de la empresa, le permite a los miembros de la empresa ver la información de forma clara y fácil de emprender para personas que no poseen un alto conocimiento sobre arquitecturas de almacenamiento o que no sean muy competentes con el uso de computadoras o de bases de datos.

### Nombres y créditos de clientes de los compradores de "Poole PLC" ordenado por crédito descendente y nombre ascendente

```python
%%sql
SELECT cliente.nombre AS nombres, cliente.credito AS credito
FROM compras
INNER JOIN cliente ON cliente.id = compras.cliente_id
WHERE tienda_id = (SELECT id from tienda where tienda.nombre = 'Poole PLC')
ORDER BY credito DESC, nombres ASC;
```

### Agregar en el resultado anterior el crédito en dólares a un tipo decambio de 7.8

```python
%%sql
SELECT cliente.nombre AS nombres, (cliente.credito / 7.8) AS credito
FROM compras
INNER JOIN cliente ON cliente.id = compras.cliente_id
WHERE tienda_id = (SELECT id from tienda where tienda.nombre = 'Poole PLC')
ORDER BY credito DESC, nombres ASC;
```

### Parejas de clientes con el mismo crédito que no sean ellos mismos y que no haya parejas duplicadas

```python
%%sql
SELECT DISTINCT a.nombre, b.nombre
FROM (cliente)a, (cliente)b 
WHERE a.credito = b.credito AND a.nombre != b.nombre AND a.nombre > b.nombre;
```

### Clientes con los creditos más altos

#### Usando max 

```python
%%sql
SELECT nombre, credito
FROM cliente WHERE credito == (SELECT MAX(credito) from cliente);
```

#### Sin usar max

```python
%%sql
SELECT nombre, credito
FROM cliente
ORDER BY credito DESC
LIMIT 10;
```

### Clientes cuyo credito en dolares sea mayor a 1

```python
%%sql
SELECT nombre, (credito/7.8) AS dolares
FROM cliente
WHERE dolares > 1;
```

### como todos tienen mas de un dolar, puse uno que no

```python
%%sql
SELECT nombre, (credito/7.8) AS dolares
FROM cliente
WHERE dolares < 1;
```

### Personas que no han hecho ninguna compra

```python
%%sql
SELECT nombre
FROM cliente
WHERE NOT EXISTS (
	SELECT *  
	FROM compras
	WHERE cliente.id = compras.cliente_id
);
```

### Número de clientes de cada tienda, de las que tienen compras, ordenado por nombre de tienda.

```python
%%sql
select tienda.nombre, count(tienda_id) as numero_clientes
FROM compras
INNER JOIN tienda ON tienda.id = compras.tienda_id
GROUP BY tienda.nombre
ORDER BY tienda.nombre ASC;
```

### Cree un trigger que considere apropiado para este esquema. Justifique su elección (5)

Se creó un trigger para restarle credito al usuario cada vez que se ingresa una compra para saber su credito real

```python
%%sql
CREATE TRIGGER descontar BEFORE INSERT
ON compras
BEGIN
	UPDATE cliente
	SET credito = credito - new.total
	WHERE id = new.cliente_id;
END;
```
